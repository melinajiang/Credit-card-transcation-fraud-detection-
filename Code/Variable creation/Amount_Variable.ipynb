## Create Attributes

```python
# Cardnum
# Merchnum
# Merch description(part)
# Cardnum-Merchnum
# Cardnum-Merch description(part)
# Cardnum-Merch state
# Cardnum-Merch zip
# Cardnum-Merch state-Merch zip

df["cardnum-merchnum"] = df["Cardnum"]+df["Merchnum"]
df["cardnum-merch_description_part"] = df["Cardnum"] + df["Merch_description_part"]
df["cardnum-merch_state"] = df["Cardnum"] + df["Merch_state"]
df["cardnum-merch_zip"] = df["Cardnum"] + df["Merch_zip"]
df["cardnum-merch_state_merch_zip"] = df["Cardnum"] + df["Merch_state"] + df["Merch_zip"]
```

```python
# count total
tempGroupby = tempdf4[["Recnum","Amount"]].groupby("Recnum").agg(["sum","mean","max","median"])
tempGroupby.columns = tempGroupby.columns.droplevel(0)
```

```python
tempGroupby
```

```python
tempGroupby = pd.merge(tempGroupby,amount_tempt,left_index = True, right_index = True, how = "left")
```

```python
tempGroupby["actual_sum"] = tempGroupby["Amount"]/tempGroupby["sum"]
tempGroupby["actual_avg"]= tempGroupby["Amount"]/tempGroupby["mean"]
tempGroupby["actual_max"]= tempGroupby["Amount"]/tempGroupby["max"]
tempGroupby["actual_median"]= tempGroupby["Amount"]/tempGroupby["median"]
```
